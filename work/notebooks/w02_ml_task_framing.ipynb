## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
"""
My project is formulated as a supervised multiclass classification problem. Using historical search intelligence signals aggregated over previous observation windows, the model predicts the future performance state of a content page. Each page will be classified into one of four business-oriented categories: Growth, Recovery, Stable, or Decline. A classification approach is appropriate because the objective is to support content prioritization decisions by assigning each page to a clearly defined future performance state rather than predicting an exact numerical change.
"""
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
"""
The final target variable will be the future content performance state, derived from the warehouse by comparing historical and future observation windows. Labels such as Growth, Recovery, Stable, and Decline will be created using predefined business rules based on changes in future search performance metrics, ensuring that only future information is used to define the target.

During the initial exploration phase, the starter dataset's trend_direction column will be used as a proxy target for exploratory data analysis and preliminary modeling. However, the final capstone will replace this proxy with labels generated directly from future warehouse observations to ensure that the model performs genuine forecasting rather than classifying already observed outcomes.
"""
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
"""
The primary evaluation metric will be Macro F1-score because the target classes are expected to be imbalanced. Macro F1-score evaluates model performance equally across all classes, preventing the majority class from dominating the evaluation. In addition to Macro F1-score, class-wise precision, recall, and a confusion matrix will be used to understand how well the model distinguishes between Growth, Recovery, Stable, and Decline. A model that achieves a strong Macro F1-score while maintaining balanced performance across all classes will be considered successful.
"""
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
"""
The fundamental unit of analysis is one content page represented over a historical observation window.

In the starter dataset, each row corresponds to a single pseudonymized content page with aggregated search performance, engagement, freshness, and content-related features. These include variables such as search_volume, impressions_90d, clicks_90d, ctr, avg_position, engagement_rate, content_age_days, and trend_direction.

For the final capstone, query-level records from the warehouse will be aggregated into page-level feature vectors over historical time windows. Each training example will therefore represent one content page summarized over a fixed observation period, while the corresponding label will be generated from a separate future evaluation window to avoid temporal leakage.
"""
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
"""
Future content performance is influenced by the interaction of many search intelligence signals rather than a single metric or threshold. A page with declining impressions may still recover if it has improving search position, strong user engagement, or increasing visibility across multiple queries. Similarly, pages with comparable click-through rates may experience different future outcomes because of differences in search demand, competition, content freshness, or historical momentum.

These relationships are nonlinear and involve complex interactions between multiple variables, making them difficult to capture with manually defined rules. Machine learning models can learn these patterns directly from historical data and combine many weak signals into more accurate predictions. As a result, an ML-based approach is better suited for forecasting future content performance and supporting evidence-based content prioritization than a fixed rule-based system.
"""
```
